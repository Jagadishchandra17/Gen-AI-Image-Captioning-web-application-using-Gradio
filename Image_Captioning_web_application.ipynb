```python
import gradio as gr
import numpy as np
from PIL import Image
from transformers import AutoProcessor, BlipForConditionalGeneration


processor = AutoProcessor.from_pretrained("Salesforce/blip2-opt-2.7b")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip2-opt-2.7b",ignore_mismatched_sizes=True)

def caption_image(input_image: np.ndarray):
    # Convert to PIL and ensure RGB
    raw_image = Image.fromarray(input_image).convert('RGB')

    # Preprocess the image
    inputs = processor(images=raw_image, return_tensors="pt")

    # Generate caption (with beam search for better results)
    generated_ids = model.generate(**inputs, max_new_tokens=50)

    # Decode output tokens into text
    caption = processor.batch_decode(generated_ids, skip_special_tokens=True)[0].strip()

    return caption

iface = gr.Interface(
    fn=caption_image,
    inputs=gr.Image(label="Upload an image"),
    outputs="text",
    title="🖼️ Image Captioning with BLIP-2",
    description="Upload any image and this app will generate a caption using the BLIP-2 model by Salesforce."
)

iface.launch()
```
